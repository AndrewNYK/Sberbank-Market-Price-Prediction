```python
import numpy as np
import pandas as pd

import seaborn as sns
```

```python
output_path = 'processed/'
```

```python
df_train = pd.read_csv('../../Dataset/train.csv/train.csv')
df_test = pd.read_csv('../../Dataset/test.csv/test.csv')

df_train.head()
```

```python
sns.boxplot(data=df_train, x='product_type', y='price_doc')
```

```python
df_train_inv = df_train[df_train['product_type'] == "Investment"]
df_train_own = df_train[df_train['product_type'] == "OwnerOccupier"]

df_test_inv = df_test[df_test['product_type'] == "Investment"]
df_test_own = df_test[df_test['product_type'] == "OwnerOccupier"]

df_train_inv.shape, df_train_own.shape, df_test_inv.shape, df_test_own.shape
```

```python
df_train['price_doc'] = df_train['price_doc'] * 0.969 + 10
df_train_inv['price_doc'] =  df_train_inv['price_doc'] * 1.05
df_train_own['price_doc'] =  df_train_own['price_doc'] * 0.9
```

```python
df_train.to_csv(output_path + "train.csv", index=False)
df_train_inv.to_csv(output_path + 'train_inv.csv', index=False)
df_train_own.to_csv(output_path + 'train_own.csv', index=False)

df_test_inv.to_csv(output_path + 'test_inv.csv', index=False)
df_test_own.to_csv(output_path + 'test_own.csv', index=False)
```
